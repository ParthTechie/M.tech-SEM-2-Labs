```python
import cv2
import matplotlib.pyplot as plt

# Path to the image
image_path = r"C:\Users\cl501_29\Desktop\Vamsidhar\MV\new-york-city-empire-state-building-panoramas-manhattan-wallpaper.jpg"

# Read the image using OpenCV
image = cv2.imread(image_path)

# Check if the image was successfully loaded
if image is not None:
    # Convert BGR to RGB
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    
    # Display the image in Jupyter Notebook
    plt.figure(figsize=(10, 6))  
    plt.imshow(image_rgb)
    plt.axis("off")  
    plt.title("Image in RGB")  
    plt.show()
else:
    print(f"Error: Unable to read the image from {image_path}")
```

```python
import cv2
import matplotlib.pyplot as plt

# Path to the image
image_path = r"C:\Users\cl501_29\Desktop\Vamsidhar\MV\new-york-city-empire-state-building-panoramas-manhattan-wallpaper.jpg"

# Read the image using OpenCV
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Check if the image was successfully loaded
if image is not None:
    
    # Display the image in Jupyter Notebook
    plt.figure(figsize=(10, 6))
    plt.imshow(image, cmap='gray')
    plt.axis("off")
    plt.title("Image in Grayscale")
    plt.show()
else:
    print(f"Error: Unable to read the image from {image_path}")
```

```python
# Cropping the image size using ratio

image_path = r"C:\Users\cl501_29\Desktop\Vamsidhar\MV\new-york-city-empire-state-building-panoramas-manhattan-wallpaper.jpg"
image = cv2.imread(image_path)

if image is not None:
    height, width, _ = image.shape

# Define the region to be cropped
top = height // 4
bottom = 3 * width // 4
left = width // 4
right = 3 * width // 4

# crop the image
cropped_image = image[top:bottom, left:right]
# Display the cropped image

if image is not None:
    
    # Display the image in Jupyter Notebook
    plt.figure(figsize=(10, 6))
    plt.imshow(cropped_image)
    plt.axis("off")
    plt.title("Cropped Image")
    plt.show()
else:
    print(f"Error: Unable to read the image from {image_path}")
```
